# Import data

```python
import pandas as pd
import numpy as np
pd.options.display.max_columns = 200
data = pd.read_csv("cleaned.csv", encoding="utf-8")
data.head(5)
```

# Analyze

NOTE: It takes too long to train model for such an extensive dataset so we build regression model using only the first 50 highest correlation features.

```python
features = data.corr()['overall'].sort_values(ascending = False)[:50]
features = features.drop('overall')
features = features.to_frame().index.tolist()
```

Split data to train and test set.

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(data,test_size=0.2,random_state=42)

X_train = train_set[features]
Y_train = train_set['overall']
X_test = test_set[features]
Y_test = test_set['overall']
```

```python
# Visualize the first 20 features vs overall score. (Maximum 20 figures)
import matplotlib.pyplot as plt
%matplotlib inline
for f in features[:20]:
    data.plot.scatter(x="overall",y=f,title="Overrall vs "+f)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()

lin_reg.fit(X_train,Y_train)
lin_reg_predictions = lin_reg.predict(X_test)
```

We compute a cross validation to produce an average rms error, which gives us a better feel for the actual model error for our model.

```python
from sklearn.model_selection import cross_val_score
shuffled_indices = np.random.permutation(len(data))
shuffled_data = data.iloc[shuffled_indices]
full_X = shuffled_data[features]
full_Y = shuffled_data['overall']
scores = cross_val_score(lin_reg,full_X,full_Y,scoring='neg_mean_squared_error',cv=5)
rms_scores = np.sqrt(-scores)
rms = rms_scores.mean()
rms
```

The root mean square error of the predictions from the model is pretty small in scale of 100. However, we will use other machine learning technique to build the efficient model to get a better result.

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=2, include_bias=False)
poly_reg = LinearRegression()
X_train_poly = poly_features.fit_transform(X_train)
poly_reg.fit(X_train_poly, Y_train)
X_test_poly = poly_features.fit_transform(X_test)
predictions = poly_reg.predict(X_test_poly)
```

```python
# Cross validation
shuffled_indices = np.random.permutation(len(data))
shuffled_data = data.iloc[shuffled_indices]
full_X = shuffled_data[features]
full_Y = shuffled_data['overall']
scores = cross_val_score(poly_reg,full_X,full_Y,scoring='neg_mean_squared_error',cv=5)
rms_scores = np.sqrt(-scores)
pr_rms = rms_scores.mean()
pr_rms
```

The mean rms error shows that the polynomial regression model does not improve the result much. Moreover, this model comsumes the most resource to run.

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

dt_reg = DecisionTreeRegressor(max_depth=2)
dt_reg.fit(X_train, Y_train)
dec_tree_predictions = dt_reg.predict(X_test)
```

```python
# Cross validation
shuffled_indices = np.random.permutation(len(data))
shuffled_data = data.iloc[shuffled_indices]
full_X = shuffled_data[features]
full_Y = shuffled_data['overall']
scores = cross_val_score(dt_reg,full_X,full_Y,scoring='neg_mean_squared_error',cv=5)
rms_scores = np.sqrt(-scores)
dt_rms = rms_scores.mean()
dt_rms
```

The root mean square error of the predictions from the model is big. Therefore, Decision Tree model does not produce good result, then it is not suitable for this dataset. 

# K-Nearest Neighbors Regression (KNN)

```python
from sklearn.neighbors import KNeighborsRegressor
knn_reg = KNeighborsRegressor(n_neighbors=5)
knn_reg.fit(X_train, Y_train) 
bag_tree_predictions = knn_reg.predict(X_test)
```

```python
# Cross validation
shuffled_indices = np.random.permutation(len(data))
shuffled_data = data.iloc[shuffled_indices]
full_X = shuffled_data[features]
full_Y = shuffled_data['overall']
scores = cross_val_score(knn_reg,full_X,full_Y,scoring='neg_mean_squared_error',cv=5)
rms_scores = np.sqrt(-scores)
knn_rms = rms_scores.mean()
knn_rms
```

By using KNN method, the mean rms error drops significantly. This model also takes less time compare to the previous one. Therefore, we conclude that KNN model is the most suitable model for this particular regression problem.
